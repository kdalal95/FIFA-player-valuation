```python
from bs4 import BeautifulSoup 
from urllib.request import urlopen as ureq
import requests
import numpy as np
import pandas as pd
import re
```

```python
players=[]
for page in range(1,101):
    Webpage_1= requests.get("https://www.fifaindex.com/players/{}/".format(page)).text
    page_soup = BeautifulSoup(Webpage_1,'lxml')
    
    table= page_soup.findAll(class_= 'table table-striped table-players')
    for tr in table:
        rows= tr.findAll('tr')
        name= tr.findAll('td',{"data-title":"Name"})
    for link in name:
        links= str(link.find('a')['href'])
        players.append(links)
        
print(players)
```
